```python
#12
file1=open('hightemp.txt','r')
col1=open('col1.txt','w')
col2=open('col2.txt','w')

for line in file1:
    #ファイルの1行ごとにタブで分けてリストにし、0行目をcol1に書き出して改行
    col1.write(line.split('\t')[0]+'\n')
    col2.write(line.split('\t')[1]+'\n')

col1.close()
col2.close()
file1.close()
```

```python
#13
col1=open('col1.txt','r')
col2=open('col2.txt','r')
col3=open('col3.txt','w')

for word1,word2 in zip(col1,col2):
    col3.write(word1.replace('\n','\t')+word2)  #word1,word2の後には\nが付いてる

col1.close()
col2.close()
col3.close()
```

```python
#18
from operator import itemgetter

file1=open('hightemp.txt','r')

#テキストファイル を１行ずつ取り出し、リスト化
file1_line=[]
for line in file1:
    line=line.strip().replace('\t',' ')
    file1_line.append(line)

#単語ごとに分ける
for i in range(0,len(file1_line)):
    file1_line[i]=file1_line[i].split()

#３コラム目の数値の降順ソート
file1_line.sort(key=itemgetter(2),reverse=True)

#リストを再び連結して表示する
for i in range(0,len(file1_line)):
    file1_line[i]=' '.join(file1_line[i])
    print(file1_line[i])
    
file1.close()
```

```python
#19
from collections import Counter
file1=open('hightemp.txt','r')

#各行の１コラム目の文字列を抽出したリストword1を作る
word1=[]
for line in file1:
    word=line.split()
    word1.append(word[0])

#word1を出現頻度順に並べ替えたlistを作る
list=Counter(word1).most_common()

#listを一つ一つ表示する
for x in list:
    print(x)

file1.close()
```
