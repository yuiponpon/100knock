```python
#05
sentense="I am an NLPer"
def ngram(sentense,n,type):
    no_comma=sentense.replace(","," ")
    no_comma_no_period=no_comma.replace("."," ")
    no_space=sentense.replace(" ","")
    
    if type == "word":
        word_ans=[]
        s1=no_comma_no_period.split()
        
        for i in range(0,len(s1)-(n-1)):
            ans1=""
            for j in range(i,i+n):
                if j==0:
                    ans1+=s1[j]
                else:
                    ans1+=" "+s1[j]
            word_ans.append(ans1)                        
        return word_ans
    
    if type == "letter":
        letter_ans=[]  
        s2=list(no_space)
        for i in range(0,len(s2)-(n-1)):
            ans2=""
            for j in range(i,i+n):
                ans2+=s2[j]
            letter_ans.append(ans2)
        return letter_ans
        
print("data1=>")
print(ngram(sentense,2,"word"))
print("data2=>")
print(ngram(sentense,2,"letter"))
```

```python
#06
s1="paraparaparadise"
s2="paragraph"
def ngram(sentense,n,type):
    no_comma=sentense.replace(","," ")
    no_comma_no_period=no_comma.replace("."," ")
    no_space=sentense.replace(" ","")
    
    if type == "word":
        word_ans=[]
        s1=no_comma_no_period.split()
        
        for i in range(0,len(s1)-(n-1)):
            ans1=""
            for j in range(i,i+n):
                if j==0:
                    ans1+=s1[j]
                else:
                    ans1+=" "+s1[j]
            word_ans.append(ans1)                        
        return word_ans
    
    if type == "letter":
        letter_ans=[]  
        s2=list(no_space)
        for i in range(0,len(s2)-(n-1)):
            ans2=""
            for j in range(i,i+n):
                ans2+=s2[j]
            letter_ans.append(ans2)
        return letter_ans
        
X=ngram(s1,2,"letter")
Y=ngram(s2,2,"letter")


X_Y_or=set(X) | set(Y)
print('X_Y_or=>',X_Y_or)

X_Y_and=set(X) & set(Y)
print('X_Y_and=>',X_Y_and)

X_Y_minus=set(X)-set(Y)
print('X_Y_minus=>',X_Y_minus)

print('"se" in X',"se" in X)

print('"se" in Y',"se" in Y)
```

```python
#08
def cipher(sentense):   
    for i in range(0,len(sentense)):      
        for j in range(ord('a'),ord('z')+1):
            if chr(j)==sentense[i]:
                sentense=sentense[:i]+chr(219-ord(sentense[i]))+sentense[i+1:]
                break
    return sentense

text="I love you."
print(cipher(text))
print(cipher(cipher(text)))
```

```python
#09
import random

sentense="I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
word_list=sentense.split()

word_list_ans=[]
for i in range(0,len(word_list)):
    if len(word_list[i])<=4:
        word_list_ans.append(word_list[i])
    else:
        word_list_ans.append(word_list[i][0]+''.join(random.sample(word_list[i][1:len(word_list[i])-1],len(word_list[i])-2))+word_list[i][len(word_list[i])-1])

ans=' '.join(word_list_ans)
print(ans)
```
